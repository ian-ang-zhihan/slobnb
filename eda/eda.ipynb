## Price Seasonality

```python
listings_dec = pd.read_csv(data_dir + "31dec2024listings.csv")
listings_dec["snapshot_date"] = "2024-12-31"

listings_mar = pd.read_csv(data_dir + "28march2025listings.csv")
listings_mar["snapshot_date"] = "2025-03-28"

listings_jun = listings_clean.copy()
listings_jun["snapshot_date"] = "2025-06-28"
```

```python
temporal_cols = ["id", "price", "room_type", "accommodates", "bedrooms", "beds", "neighbourhood_cleansed", "snapshot_date"]

# Combine all available snapshots
temporal_data = pd.concat(
    [
        listings_dec[temporal_cols],
        listings_mar[temporal_cols],
        listings_jun[temporal_cols]
    ],
    ignore_index=True
)

print(f"Combined temporal dataset: {len(temporal_data)} rows")
print(f"Unique listings: {temporal_data["id"].nunique()}")
print(f"Snapshots: {temporal_data["snapshot_date"].unique()}")

temporal_data.head(10)
```

```python
temporal_data["clean_price"] = temporal_data["price"].str.replace("$", "").str.replace(",", "").astype(float)

print(f"Before outlier removal: {len(temporal_data)}")
temporal_data["snapshot_date"] = pd.to_datetime(temporal_data["snapshot_date"])

temporal_data_clean = temporal_data[
    (temporal_data["clean_price"] >= 0) &
    (temporal_data["clean_price"] < 50000)
].copy()

temporal_data = temporal_data_clean
print(f"After outlier removal: {len(temporal_data)}")
```

```python
num_snapshots = temporal_data["snapshot_date"].nunique()
listing_counts = temporal_data.groupby("id").size()
stable_listings = listing_counts[listing_counts == num_snapshots].index

stable_data = temporal_data[temporal_data["id"].isin(stable_listings)].copy()
stable_data = stable_data.sort_values(["id", "snapshot_date"])
stable_data
```

```python
avg_prices = stable_data.groupby("snapshot_date")["clean_price"].agg(["mean", "median", "count"])
print(avg_prices)
```

```python
avg_prices[["mean", "median"]].plot.line(
    xlabel="Date",
    ylabel="Price ($)",
    title="Avg Listing Price Over Time",
    marker="o"
)
```

```python
room_type_trends = stable_data.groupby(["snapshot_date", "room_type"])["clean_price"].median().unstack()
print(room_type_trends)
```

```python
room_type_trends.plot.line(
    xlabel="Date",
    ylabel="Median Price ($)",
    title="Price Trends by Room Type Over Time",
    marker="o"
)
```

```python
import random
random.seed(42)

sample_ids = random.sample(list(stable_listings), min(10, len(stable_listings)))
print(sample_ids)

sample_data = stable_data[stable_data["id"].isin(sample_ids)]
# print(sample_data)

price_trajectories = sample_data.pivot(
    index="snapshot_date",
    columns="id",
    values="clean_price"
)

price_trajectories.plot.line(
    xlabel="Date",
    ylabel="Price ($)",
    title="Price Trajectories for 10 Samples",
    marker="o",
    alpha=0.7
)
plt.legend(title="Listing ID", bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=8)
```
